```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed dataset
df = pd.read_csv("processed_for_peak_analysis.csv")

# Function to analyze a player's career trajectory with separate plots
def analyze_player_trajectory(player_name, position, rolling_window=5):
    """
    Generates multiple separate performance trajectory graphs for a given player.
    Each graph corresponds to one key performance metric.
    - rolling_window: Defines how many seasons should be averaged for trend smoothing.
    """

    # Filter data for the given player
    player_df = df[df["player"] == player_name].sort_values("season")

    # Extract relevant metrics based on position
    position_metrics = {
        "FW": ["Goals p 90", "Expected Goals (xG)", "% Shots on target", "Total Shots",
               "Progressive Carries", "Goal Creating Actions p 90", "Shot Creating Actions p 90"],
        "MF": ["Assists p 90", "Key passes", "Progressive Passes", "Pass completion %",
               "Tackles Won", "Progressive Carries", "Goal Creating Actions p 90"],
        "DF": ["Tackles Won", "Interceptions", "% Aerial Duels won", "Clearances",
               "Errors made", "Pass completion %"],
        "GK": ["Saves %", "Clean Sheets", "Goals against p 90", "% Penalty Saves", "Crosses Stopped"]
    }

    metrics = position_metrics.get(position, ["Goals p 90"])  # Default to Goals p 90 if missing

    # Create separate plots for each metric
    for metric in metrics:
        if metric in player_df.columns:
            player_df[f"rolling_{metric}"] = player_df[metric].rolling(window=rolling_window, min_periods=1).mean()

            # Identify peak season (highest rolling average)
            peak_season = player_df.loc[player_df[f"rolling_{metric}"].idxmax(), "season"]
            peak_value = player_df[f"rolling_{metric}"].max()

            # Create a new figure for each metric
            plt.figure(figsize=(10, 5))
            sns.lineplot(x=player_df["season"], y=player_df[f"rolling_{metric}"], marker='o', linestyle='-', label=f"{metric}")

            # Highlight peak season
            plt.scatter(peak_season, peak_value, color='red', s=100, label=f"Peak Season ({metric})")

            # Styling
            plt.title(f"Career Performance Trajectory - {player_name} ({position}) - {metric}")
            plt.xlabel("Season")
            plt.ylabel(metric)
            plt.legend()
            plt.grid()
            plt.show()

# Example usage: Analyze a player's career trajectory with separate plots
analyze_player_trajectory(player_name="Raheem Sterling", position="FW", rolling_window=5)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import Dataset, DataLoader

# Load dataset
df = pd.read_csv("processed_for_peak_analysis.csv")

# Ensure season is treated as an integer
df["season"] = df["season"].astype(int)

# Define the metric for prediction (e.g., Goals per 90 minutes)
metric = "Goals p 90"  # Change this to any relevant metric
player_name = "Robert Lewandowski"  # Change this to any player

# Filter the data for the player
player_df = df[df["player"] == player_name].sort_values(by="season")

# Ensure the player has enough historical data
if len(player_df) < 3:
    raise ValueError(f"Not enough data to train LSTM for {player_name}")

# Extract the seasons and metric values
seasons = player_df["season"].values
metric_values = player_df[metric].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler()
metric_values_scaled = scaler.fit_transform(metric_values)

# Prepare sequences for LSTM (Sliding Window Approach)
sequence_length = 3  # Using last 3 seasons to predict the next

X, y = [], []
for i in range(len(metric_values_scaled) - sequence_length):
    X.append(metric_values_scaled[i:i + sequence_length])
    y.append(metric_values_scaled[i + sequence_length])

X = np.array(X)
y = np.array(y)

# Convert to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Define Dataset and DataLoader
class PlayerDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = PlayerDataset(X_tensor, y_tensor)
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

# Define LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=2, output_size=1):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# Initialize Model, Loss, and Optimizer
model = LSTMModel()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Train the LSTM Model
epochs = 500
for epoch in range(epochs):
    for batch_X, batch_y in dataloader:
        optimizer.zero_grad()
        output = model(batch_X)
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")

# Predict for the next season
X_test = torch.tensor(metric_values_scaled[-sequence_length:].reshape(1, sequence_length, 1), dtype=torch.float32)
predicted_scaled = model(X_test).detach().numpy()

# Convert prediction back to original scale
predicted_value = scaler.inverse_transform(predicted_scaled)[0][0]
actual_last_season = metric_values[-1][0]

# Print results
print(f"\n🔮 Predicted {metric} for {player_name} in {seasons[-1] + 1}: {predicted_value:.4f}")
print(f"📊 Actual {metric} in {seasons[-1]}: {actual_last_season:.4f}")

# Plot Actual vs. Predicted
plt.figure(figsize=(8, 5))
plt.plot(seasons, metric_values, marker="o", linestyle="-", label="Actual", color="blue")
plt.scatter([seasons[-1] + 1], [predicted_value], color="red", label="Predicted", marker="o", s=100)
plt.xlabel("Season")
plt.ylabel(metric)
plt.title(f"LSTM Performance Prediction - {player_name}")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
df = pd.read_csv("/content/cleaned_all_seasons.csv")

# Encode league as numeric
df["league_index"] = LabelEncoder().fit_transform(df["comp"])

# Select all numeric features (excluding season)
feature_columns = df.select_dtypes(include=[np.number]).columns.tolist()
feature_columns.remove("season")

# Fill missing values
df[feature_columns] = df[feature_columns].fillna(df[feature_columns].median())

# Scale numeric features
scaler = StandardScaler()
df[feature_columns] = scaler.fit_transform(df[feature_columns])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor

# Assuming df is already loaded and cleaned
positions = df["pos"].unique()
feature_importance_dict = {}

for pos in positions:
    print(f"\n🔍 Analyzing most important stats for {pos}...\n")

    # Filter data for position
    pos_df = df[df["pos"] == pos]

    # Ensure sufficient data
    if len(pos_df) < 50:
        print(f"⚠️ Not enough data for {pos}. Skipping...\n")
        continue

    # Use all numerical columns except non-performance ones
    exclude_cols = ["rk", "player", "nation", "pos", "squad", "comp", "season", "key_metrics"]
    feature_columns = [col for col in df.columns if col not in exclude_cols]

    X = pos_df[feature_columns]  # Use all stats as features
    y = np.random.rand(len(X))   # Placeholder target (we only care about feature ranking)

    # Train Random Forest Model (No real target, just ranking)
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X, y)

    # Get feature importance
    importance = model.feature_importances_
    feature_importance = dict(zip(feature_columns, importance))

    # Sort features by importance
    sorted_features = sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)[:15]

    # Store only top 15
    feature_importance_dict[pos] = dict(sorted_features)

    # Plot feature importance
    plt.figure(figsize=(10, 6))
    sns.barplot(x=[imp[1] for imp in sorted_features], y=[imp[0] for imp in sorted_features], palette="viridis")
    plt.title(f"Top 15 Feature Importance for {pos}")
    plt.xlabel("Importance Score")
    plt.ylabel("Feature")
    plt.show()
```

```python
# Convert feature importance to DataFrame
feature_importance_df = pd.DataFrame(feature_importance_dict).fillna(0)

# Get top 10 features for each position
top_features_per_position = feature_importance_df.apply(lambda x: x.nlargest(15).index.tolist(), axis=0)
print("\n🔹 Top 15 Important Features for Each Position:\n")
print(top_features_per_position)
```
